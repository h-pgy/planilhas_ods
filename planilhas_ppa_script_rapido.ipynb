```python
import os
import pandas as pd
from openpyxl import load_workbook
```

```python
data_path = 'original_data/Indicadores - planilhas recebidas SF/'
```

```python
files = [os.path.join(data_path, file) for file in os.listdir(data_path)
        if not file.endswith('#')]
```

```python
devolutivas = 'DEVOLUTIVA IndicadoresProgramas'
preencher = 'PREENCHER Indicadores Ações'
```

```python
dfs_devolutivas = [pd.read_excel(file, sheet_name=devolutivas) for file in files]
```

```python
df_final_devolutivas = pd.concat(dfs_devolutivas)
```

```python
def drop_unanamed_cols(df):
    
    df = df.copy()
    
    dropar = []
    for col in df.columns:
        if str(col).startswith('Unnamed: '):
            dropar.append(col)
            
    df.drop(dropar, axis = 1, inplace=True)
    
    return df
```

```python
df_final_devolutivas = drop_unanamed_cols(df_final_devolutivas)
```

```python
df_final_devolutivas.to_excel('devolutivas_unificadas.xlsx')
```

```python
dfs_preencher = [pd.read_excel(file, sheet_name=preencher, skiprows=1) for file in files]
```

```python
df_final_preencher=  pd.concat(dfs_preencher)
```

```python
df_final_preencher = drop_unanamed_cols(df_final_preencher)
```

```python
df_final_preencher
```

```python
linhas_vazias = df_final_preencher.isnull().all(axis=1)
```

```python
df_final_preencher[linhas_vazias]
```

```python
df_final_preencher = df_final_preencher[~linhas_vazias].copy()
```

```python
df_final_preencher.to_excel('preenchidas_unificadas.xlsx')
```
